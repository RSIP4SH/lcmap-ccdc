```python
from cytoolz     import take

from firebird  import ARD
from firebird  import grid
from firebird  import pyccd
from firebird  import timeseries
from firebird  import ids
from functools import partial

from importlib   import reload
from merlin      import functions as f
from pyspark.sql import functions as ff
from pyspark.sql import SparkSession
from pyspark.sql import types

import firebird
import pyspark
```

```python
firebird.ARD
```

```python
ss = SparkSession(pyspark.SparkContext.getOrCreate())
```

```python
# point in tile 005015
x = -1738207
y = 992001

tile = grid.tile(x=x, y=y, cfg=ARD)

chips = take(1, tile.get('chips'))
```

```python
cids = ids.rdd(ctx=ss.sparkContext, cids=chips)

#ard = timeseries.ard(ctx=ss.sparkContext, cids=cids, acquired='1980-01-01/2017-01-01')
ard = timeseries.rdd(ctx=ss.sparkContext, cids=cids, acquired='1980-01-01/2017-01-01', cfg=firebird.ARD, name='ard')
```

```python
ard
```

```python
ccdrdd = pyccd.rdd(ctx=ss.sparkContext, timeseries=ard)
```

```python
results = ccdrdd.collect()
```
